```python
import pandas as pd
import matplotlib.pyplot as plt
```

###
## Loading DataFrame:

```python
df1 = pd.read_csv('4-pr1-DataSet-living cost-2024.csv')
df1
```

###
## Creating Dataframe with only 35 Countries:
###

```python
countries_of_interest = [
    # Asia
    'Pakistan', 'China', 'Iran', 'Bangladesh', 'Japan',
    # North America
    'United States', 'Canada', 'Mexico',
    # South America
    'Brazil', 'Argentina', 'Colombia', 'Peru', 'Venezuela',
    # Europe
    'France', 'Italy', 'United Kingdom', 'Netherlands',
    # Africa
    'Nigeria', 'South Africa', 'Egypt', 'Ethiopia',
    # Australia/Oceania
    'Australia', 'New Zealand', 'Fiji', 'Tonga'
]

df = df1[df1['Country'].isin(countries_of_interest)]
df
```

###
## Line plot:
**Visualizing each column with respect to each country.**

###

```python
import numpy as np
```

```python
columns = ['Cost of Living Index', 'Rent Index', 
           'Cost of Living Plus Rent Index', 'Groceries Index',
       'Restaurant Price Index', 'Local Purchasing Power Index']
a = 0

for i in columns: 
    
    a += 1
    y = df['Country']
    x = df[i]

    plt.figure(figsize = (7, 7))

    plt.yticks(fontsize = 9)
    plt.xticks(fontsize = 14)

#    plt.plot(x, y, marker = 'o', linestyle = None, linewidth = 2, markersize = 7, markerfacecolor = 'g', markeredgecolor = 'r', markeredgewidth = 1, label = i)
    plt.plot(x, y, color = 'b', linestyle = '--', linewidth = 1, marker = 'o', markersize = 7, markerfacecolor = 'g', markeredgecolor = 'r', markeredgewidth = 2, label = 'Consumption rate in kg')

    plt.legend()

    plt.xlabel(f'\n{i}')
    plt.ylabel('Countries')
    plt.title(f'{a}) "Countries w.r.t {i}"\n')

    plt.grid(color = 'k')
```

###
## Bar Plot:
**Visualizing each column with respect to each country.**

###

```python
columns = ['Cost of Living Index', 'Rent Index', 
           'Cost of Living Plus Rent Index', 'Groceries Index',
       'Restaurant Price Index', 'Local Purchasing Power Index']
a = 0

for i in columns: 
    
    a += 1
    x = df['Country']
    y = df[i]

    plt.figure(figsize = (7, 7))

    plt.yticks(fontsize = 9)
    plt.xticks(fontsize = 14)

    plt.barh(x, y, height = 0.5)

    plt.xlabel(f'\n{i}')
    plt.ylabel('Countries')
    plt.title(f'{a}) "Countries w.r.t {i}"\n')

    plt.grid(color = 'k')
```

###
## Scatter Plot:
**Visualizing each column with respect to each country.**

###

```python
columns = ['Cost of Living Index', 'Rent Index', 
           'Cost of Living Plus Rent Index', 'Groceries Index',
       'Restaurant Price Index', 'Local Purchasing Power Index']

a = 0

for i in columns: 
    
    a += 1
    y = df['Country']
    x = df[i]

    plt.figure(figsize = (7, 7))

    plt.yticks(fontsize = 9)
    plt.xticks(fontsize = 14)

    plt.scatter(x, y, color = 'green')

    plt.xlabel(f'\n{i}')
    plt.ylabel('Countries')
    plt.title(f'{a}) "Countries w.r.t {i}"\n')

    plt.grid(color = 'k')
```

###
## Pie Chart:
**Visualizing each country's data with respect to other countries.**

###

```python
columns = ['Cost of Living Index', 'Rent Index',
       'Cost of Living Plus Rent Index', 'Groceries Index',
       'Restaurant Price Index', 'Local Purchasing Power Index']

print('Columns: ', columns)
column = str(input('\nEnter column name: '))
    
a = df[column]
b = df['Country']

plt.figure(figsize=(7, 7))
plt.pie(a, labels = b, autopct='%1.1f%%', startangle = 1, textprops={'fontsize': 7})
plt.title(f'{column}')
plt.show()
plt.tight_layout()
print(f'\nIt is showing how much each country\'s "{column}" is out of 100%.\n')
```

###
## Scatter Plot:
**Visualizing copmarisions between different columns.**:

###

```python
column1 = str(input('Enter column1 (x-axis): '))
column2 = str(input('Enter column2 (y-axis): '))
column3 = str(input('Enter column3: '))

x = df[column1]
y = df[column2]
z = df[column3]

plt.figure(figsize=(4, 4))
plt.scatter(x, y, marker = 'o', s = 10, alpha = 1, c=z, cmap='viridis')
plt.colorbar(label='Color Map Label')
plt.grid(color = 'grey')
plt.title('"Cost of Living Index" vs "Groceries Index"\n')
plt.xlabel(column1)
plt.ylabel(column2)
plt.show()

```

###
## Visualizing Expenses  vs Total Cost:
###

```python
df1 = df.set_index('Country')
```

```python
country = str(input('Enter country name: '))
column = str(input('Which column you wanna compare with with "Cost of Living": '))

a = df1.loc[country]
b = a['Cost of Living Index']
c = a[column]


labels = ['Cost of Living', column]
sizes = [b, c]

plt.pie(sizes, labels = labels, autopct='%1.1f%%', textprops={'fontsize': 12}, shadow = True, explode = (0,0.17))
plt.title(f'\nCost of Living vs {column} - {country}')
plt.show()
```

###
## Comparision between two Values:
###

```python
country = str(input('Enter country name: '))
column1 = str(input('Enter column1: '))
column2 = str(input('Enter column2: '))


a = df1.loc[country]
b = a[column1]
c = a[column2]


labels = [column1, column2]
sizes = [b, c]

plt.pie(sizes, labels = labels, autopct='%1.1f%%', textprops={'fontsize': 12}, shadow = True, explode = (0,0.17))
plt.title(f'{column1} vs {column2} - {country}')
plt.show()
```
